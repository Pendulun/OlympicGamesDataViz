# Alguns dados ao longo dos anos

```python
uniqueAthletes = athletesDf.drop_duplicates(['ID','Games'])
uniqueAthletesPerYear = uniqueAthletes.groupby('Year')
```

```python
def plotMaleFemalePerYear(uniqueAthletesPerYear, ax):
    """
    Quantidade de homens e mulheres ao longo dos anos
    """
    maleFemalePerYear = uniqueAthletesPerYear['Sex'].value_counts()
    
    males = maleFemalePerYear[maleFemalePerYear.index.get_level_values('Sex') == 'M']
    ax.plot(maleFemalePerYear.index.get_level_values('Year').unique(), males.values, label='Homens')
    females = maleFemalePerYear[maleFemalePerYear.index.get_level_values('Sex') == 'F']
    ax.plot(maleFemalePerYear.index.get_level_values('Year').unique()[1:], females.values, label='Mulheres')
    ax.legend()
    ax.set_ylabel("Quantidade")
    ax.set_title("Quantidade de homens e mulheres ao longo dos jogos")

def plotMeanHeightPerSexPerYear(uniqueAthletesPerYear, uniqueAthletes, ax):
    """
    Altura média de homens e mulheres ao longo dos anos
    """
    meanHeightPerYear = uniqueAthletesPerYear['Height'].mean()
    uniqueMaleFemalePerYear = uniqueAthletes.groupby(["Year",'Sex'])
    meanHeights = uniqueMaleFemalePerYear['Height'].mean()
    males = meanHeights[meanHeights.index.get_level_values('Sex') == 'M']
    ax.plot(meanHeightPerYear, label="Média geral")
    ax.plot(meanHeights.index.get_level_values('Year').unique(), males.values, label='Homens')
    females = meanHeights[meanHeights.index.get_level_values('Sex') == 'F']
    ax.plot(meanHeights.index.get_level_values('Year').unique()[1:], females.values, label='Mulheres')
    ax.legend()
    ax.set_ylabel("Altura Média (cm)")
    ax.set_title("Altura média de homens e mulheres ao longo dos jogos")

def plotMeanWeightPerSexPerYear(uniqueAthletesPerYear, uniqueAthletes, ax):
    """
    Peso médio de homens e mulheres ao longo dos anos
    """
    meanWeights = uniqueMaleFemalePerYear['Weight'].mean()
    meanWeightPerYear = uniqueAthletesPerYear['Weight'].mean()
    males = meanWeights[meanWeights.index.get_level_values('Sex') == 'M']
    ax.plot(meanWeightPerYear, label="Média geral")
    ax.plot(meanWeights.index.get_level_values('Year').unique(), males.values, label='Homens')
    females = meanWeights[meanWeights.index.get_level_values('Sex') == 'F']
    ax.plot(meanWeights.index.get_level_values('Year').unique()[1:], females.values, label='Mulheres')
    ax.legend()
    ax.set_ylabel("Peso Médio (Kg)")
    ax.set_title("Peso médio de homens e mulheres ao longo dos jogos")


fig, axs1 = plt.subplots(1, 1, figsize=(10,6))
plotMaleFemalePerYear(uniqueAthletesPerYear, axs1)
plt.show()

fig, axs2 = plt.subplots(1, 2, figsize=(18,7))
plotMeanHeightPerSexPerYear(uniqueAthletesPerYear, uniqueAthletes, axs2[0])
plotMeanWeightPerSexPerYear(uniqueAthletesPerYear, uniqueAthletes, axs2[1])
plt.show()
```

Em 1896, as mulheres não podiam participar. Esse gráfico inclui tanto jogos de inverno quanto jogos de verão. De acordo com o [Wikipedia](https://pt.wikipedia.org/wiki/Jogos_Ol%C3%ADmpicos_de_Inverno#1984_a_1998), até 1994 os Jogos Olímpicos de Inverno eram realizados no mesmo ano que os Jogos Olímpicos de Verão. Isso mudou em 1994, quando os jogos alternaram a cada 2 anos.
